```python
import sys, os, time, cv2, random, datetime
from glob import glob
import xml.etree.ElementTree as ET
from shutil import copyfile
from PIL import Image
from utils.utils import check_conditions, delete_content
```

```python
# Set the input directory paths : path to the dataset you need to use. We assume the datas are annotated in Pascal VOC format
INPUT_DIR_1 = "/home/sitou/Documents/2.Research/02.COVID-19/Datasets/9_MAFA"
INPUT_DIR_2 = "/home/sitou/Documents/2.Research/02.COVID-19/Datasets/10_REAL_WIDERPERSON_RFB_NOMASK"
CURRENT_WORKING_DIRECTORY = os.getcwd()

#Set the train/test images spliting ratio
TRAIN_TEST_SPLIT = 0.2

#Set the directories to use in a Python list
#data_list = [INPUT_DIR_2]
data_list = [INPUT_DIR_1, INPUT_DIR_2]

# Set the paths of your Tensorflow API models
TENSORFLOW_MODELS = "/home/sitou/Documents/2.Research/02.COVID-19/Datasets/00_Notebooks/tensorflow/research"
IMAGES_DIR        = os.path.join(TENSORFLOW_MODELS, "object_detection", "images")
TRAIN_DIR         = os.path.join(TENSORFLOW_MODELS, "object_detection", "images", "train")
TEST_DIR          = os.path.join(TENSORFLOW_MODELS, "object_detection", "images", "test")

# Start cleaning the target folders
if os.path.exists(TRAIN_DIR):
    print("Train folder exists. Cleaning it !")
    delete_content(TRAIN_DIR)
else:
    print("Train folder does not exist. Creating it !")
    os.mkdir(TRAIN_DIR)

if os.path.exists(TEST_DIR):
    print("Test folder exists. Cleaning it !")
    delete_content(TEST_DIR)
else:
    print("Test folder does not exist. Creating it !")
    os.mkdir(TEST_DIR)
```

```python
# Spliting datas : Train/Test
TRAIN_TEST_SPLIT_NUMBER = int(1/TRAIN_TEST_SPLIT)
iteration = 0
for folder in data_list:
    for image_file in os.listdir(folder):
        if image_file.split('.')[1] == "jpg":
            xml_file = image_file.split('.')[0] + '.xml'
            print("#### ", image_file)
            iteration += 1
            if iteration % TRAIN_TEST_SPLIT_NUMBER == 0:
                copyfile(os.path.join(folder, image_file), os.path.join(IMAGES_DIR, "test", image_file))
                copyfile(os.path.join(folder, xml_file), os.path.join(IMAGES_DIR, "test", xml_file))
            else:
                copyfile(os.path.join(folder, image_file), os.path.join(IMAGES_DIR, "train", image_file))
                copyfile(os.path.join(folder, xml_file), os.path.join(IMAGES_DIR, "train", xml_file))
```
